```python
%pip install pandas matplotlib seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import os
from pathlib import Path

# Paths
RAW_DATA_PATH = Path("data/raw/complaints.csv")
FILTERED_DATA_PATH = Path("data/processed/filtered_complaints.csv")

# Load the dataset
# Load the dataset from local path
df = pd.read_csv(RAW_DATA_PATH)

# ------------------------------
# 1. Initial EDA
# ------------------------------
print("Dataset shape:", df.shape)
print("Columns:\n", df.columns)
print("\nMissing values:\n", df.isnull().sum())

# Check sample rows
print("\nSample rows:\n", df[['Product', 'Consumer complaint narrative']].sample(5))

# ------------------------------
# 2. Complaints by Product
# ------------------------------
product_counts = df['Product'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(x=product_counts.index, y=product_counts.values)
plt.xticks(rotation=45, ha='right')
plt.title("Complaint Counts by Product")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ------------------------------
# 3. Narrative Length Analysis
# ------------------------------
df['narrative_length'] = df['Consumer complaint narrative'].dropna().apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 5))
sns.histplot(df['narrative_length'].dropna(), bins=50, kde=True)
plt.title("Distribution of Complaint Narrative Length (in words)")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

print("\nNarrative Length Statistics:")
print(df['narrative_length'].describe())

# ------------------------------
# 4. Count Narratives vs Non-narratives
# ------------------------------
with_narrative = df['Consumer complaint narrative'].notnull().sum()
without_narrative = df['Consumer complaint narrative'].isnull().sum()
print(f"\nComplaints with narrative: {with_narrative}")
print(f"Complaints without narrative: {without_narrative}")

# ------------------------------
# 5. Filter Dataset for Target Products
# ------------------------------
target_products = [
    'Credit card', 'Personal loan',
    'Buy Now, Pay Later', 'Savings account',
    'Money transfer, virtual currency'
]

df_filtered = df[
    df['Product'].isin(target_products) &
    df['Consumer complaint narrative'].notnull()
].copy()

# ------------------------------
# 6. Text Cleaning
# ------------------------------
def clean_text(text):
    text = text.lower()
    text = re.sub(r'i am writing to (file|submit) a complaint.*?[\.\!]', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_filtered['cleaned_narrative'] = df_filtered['Consumer complaint narrative'].apply(clean_text)

# Save cleaned dataset
df_filtered.to_csv(FILTERED_DATA_PATH, index=False)
print(f"\nFiltered and cleaned dataset saved to {FILTERED_DATA_PATH}")
```
